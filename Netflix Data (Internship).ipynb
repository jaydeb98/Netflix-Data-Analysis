# Performing EDA on the Data Set

## 1. Content Distribution Analysis on NETFLIX 

```python
freq = data.type.value_counts() #To Check the Type of Contents Available on Netflix and Store in the 'freq' Variable
```

```python
# To Create a Figure for Both Charts
plt.figure(figsize=(12, 6))

# To Plot the Bar Chart Using Seaborn
plt.subplot(1, 2, 1)
sns.barplot(x=freq.index, y=freq.values, palette='Set2')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')

# To Plot the Pie Chart Using Matplotlib
plt.subplot(1, 2, 2)
plt.pie(freq, labels=freq.index, autopct='%1.0f%%', colors=['skyblue', 'lightgreen'])
plt.title('Percentage of Movies vs TV Shows')

# Display the charts
plt.tight_layout()

# Add a super title
plt.suptitle('Netflix Content Distribution: Movies vs TV Shows', fontsize=20, y=1.05)

# Display the plots
plt.show()
```

## 2. Rating Frequency of Movies and TV Shows on NETFLIX

```python
rate = data['rating'].value_counts() #To check the different types of ratings and storing their counts on 'rate' variable
```

```python
rate 
```

```python
# To Create a Figure for Both Charts
plt.figure(figsize=(12, 6))

# To Plot the Bar Chart Using Seaborn
plt.subplot(1, 2, 1)
sns.barplot(x=rate.index, y=rate.values, palette='Set2')
plt.title('Rating on Netflix')
plt.xlabel('Rating Types')
plt.ylabel('Rating Frequency')
plt.xticks(rotation = 90)

# To Plot the Pie Chart Using Matplotlib
plt.subplot(1, 2, 2)
plt.pie(rate[:8], labels=rate[:8].index, autopct='%1.0f%%')
plt.title('Rating on Netflix')

# Display the charts
plt.tight_layout()

# Add a super title
plt.suptitle('Netflix Ratings', fontsize=20, y=1.05)

# Display the plots
plt.show()
```

## 3. Top 10 Countries with Most Content on NETFLIX

## 4. Monthly Release of Movies and TV Series on NETFLIX

```python
#To Add a Separate Month Column in the Data

data['date_added'] = pd.to_datetime(data['date_added'])

# Create a 'month' column by extracting the month from the 'release_date' column
data['month'] = data['date_added'].dt.month
```

```python
#To Create a Side-by-Side Bar Chart

monthly_movie_release = data[data['type']=='Movie']['month'].value_counts().sort_index()
monthly_series_release = data[data['type']=='TV Show']['month'].value_counts().sort_index()

monthly_release = pd.DataFrame({
    'Movies': monthly_movie_release,
    'Series': monthly_series_release
}).fillna(0)  # Fill missing months with 0

sns.set(style="whitegrid")

monthly_release.plot(kind='bar', figsize=(9, 6), color=sns.color_palette("pastel"))

# Customize the plot
plt.xlabel("Months")
plt.ylabel("Frequency of releases")
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
plt.legend()

# Add a suptitle
plt.suptitle("Monthly Releases of Movies and TV Series on Netflix")

# Disable grid
plt.grid(False)

# To display the chart
plt.show()
```

## 5. Yearly Release of Movies and TV Series on NETFLIX 

```python
#To Create a Line Chart

yearly_movie_release = data[data['type']=='Movie']['release_year'].value_counts().sort_index()
yearly_series_release = data[data['type']=='TV Show']['release_year'].value_counts().sort_index()

sns.set(style="whitegrid")

plt.figure(figsize=(9, 6)) 

plt.plot(yearly_movie_release.index, yearly_movie_release.values, label='Movies', color=sns.color_palette("pastel")[0], linewidth=2)
plt.plot(yearly_series_release.index, yearly_series_release.values, label='TV Shows', color=sns.color_palette("pastel")[1], linewidth=2)

# Adding labels and title
plt.xlabel("Years")
plt.ylabel("Frequency of releases")
plt.grid(True)
plt.suptitle("Yearly Releases of Movies and TV Series on Netflix")

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

## 6. Top 10 Popular Movie Genres

```python
#To Sort Out the Top 10 Movie Genres in Netflix
popular_movie_genre = data[data.type =='Movie'].groupby("listed_in").size().sort_values(ascending=False)[:10]

popular_movie_genre
```

```python
#To Plot the Data in Bar Chart


plt.figure(figsize=(12, 6))  # Set the figure size

# Create the bar chart
sns.barplot(x=popular_movie_genre.index, y=popular_movie_genre.values, palette="viridis")

# Adding titles and labels
plt.title('Top 10 Movie Genres on Netflix', fontsize = 16)
plt.xlabel('Movie Genres')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid lines only on y-axis

# Show the bar chart
plt.show()
```

## 7. Top 10 TV Series Genre

```python
#To Sort Out the Top 10 TV Series in Netflix
popular_tv_series = data[data.type =='TV Show'].groupby("listed_in").size().sort_values(ascending=False)[:10]

popular_tv_series 
```

```python
#To Plot the Data in Bar Chart


plt.figure(figsize=(12, 6))  

# Create the bar chart
sns.barplot(x=popular_tv_series.index, y=popular_tv_series.values, palette="plasma")

# Adding titles and labels
plt.title('Top 10 TV Series on Netflix', fontsize = 16)
plt.xlabel('Series Genres')
plt.ylabel('Number of Series')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid lines only on y-axis

# Show the bar chart
plt.show()
```
